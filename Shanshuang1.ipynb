```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('desktop/000001.csv')
data["Day"] = pd.to_datetime(data["Day"],format="%Y/%m/%d")
data = data.sort_values(by='Day',ascending=True)
data.set_index("Day",inplace=True)
data
```

```python
data['2015-06-01':'2015-06-30']
```

```python
#股指日数据收益率
```

```python
%%time
data_new = data['1995-01':'2025-08'].copy()
data_new['Close'] = pd.to_numeric(data_new['Close'],errors='coerce')
data_new['Preclose'] = pd.to_numeric(data_new['Preclose'],errors='coerce')
data_new
```

```python
data_new["Raw_return"] = data_new['Close'] / data_new['Preclose'] - 1
data_new["Log_return"] = np.log(data_new['Close'] / data_new['Preclose'])
data_new["Pct_change"] = data_new['Close'].pct_change()
data_new
```

```python
#月度收益率
#"M"是月末（month end） “MS”是月初（month start）
Month_data = data_new.resample('M')['Close'].last().to_frame()
Month_data['Preclose'] = Month_data['Close'].shift(1)
Month_data['Month_return'] = Month_data['Close'] / Month_data['Preclose'] - 1
Month_data.dropna(inplace=True)
Month_data
```

```python
Month_data = data_new.resample('M')['Raw_return'].apply(lambda x: (x + 1).prod() - 1).to_frame()
Month_data.rename(columns={'Raw_return':'Month_return'},inplace=True)
Month_data.dropna(inplace=True)
Month_data
```

```python
#年度收益率&平均值
yearly_close = data_new['Close'].resample('Y').last()
Year_data = yearly_close.pct_change().to_frame(name='Year_return')
avg_year_return = np.mean(Year_data['Year_return'])
print(avg_year_return)
```
